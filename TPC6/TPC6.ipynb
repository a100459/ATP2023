Crie um programa em Python, conjunto de funções, que responda às seguintes questões:
* Crie uma função que lê a informação do ficheiro para um modelo, previamente pensado em memória;
* Crie uma função que calcula a distribuição da doença por sexo (tenha atenção que indivíduos doentes e não doentes no dataset);
* Crie uma função que calcula a distribuição da doença por escalões etários. Considere os seguintes escalões: [0-10], [11-24], [25-29], [30-34], [35-39], [40-44], ...
* Crie uma função que calcula a distribuição da doença por níveis de glucose. Considere um nível igual a um intervalo de 10 unidades, comece no limite inferior e crie os níveis necessários até abranger o limite superior;
* Crie uma função que imprime na forma de uma tabela uma distribuição;
* Especifique um programa que ao executar apresenta as tabelas correspondentes às distribuições pedidas.

```python
def lerFicheiro(fnome):
    dados = []
    ficheiro = open(fnome)
    for line in ficheiro:     
        paciente=line[:-1].split(",")
        dados.append(paciente)
    ficheiro.close()
    return dados 

lerFicheiro("diabetes_prediction_dataset.csv")

def distrib_Sexo(lista):
    masculino = 0
    feminimo = 0
    for p in lista:
        if p[8] == "1":
            if p[0] == "Male":
                masculino = masculino + 1
            if p[0] == "Female":
                feminimo = feminimo + 1
    distrib=[("Male",masculino),("Female",feminimo)]
    return distrib

def distrib_Idades(lista):
    idade_Max=0
    for i in lista:
        if float(i[1]) > idade_Max:
            idade_Max = float(i[1])
    while idade_Max % 5 != 0:
        idade_Max += 1
    escaloes = [[(0,10),0], [(11,24),0]]+[[(x, x+4), 0] for x in range(25, int(idade_Max), 5)]
    for i in lista:
        if i[-1] == "1":
            idade= float(i[1])
            if idade<=10:
                if escaloes[0][0][0]<= idade and idade<=escaloes[0][0][1]:
                    escaloes[0][1] += 1
            if idade<25 and idade>10:
                if escaloes[1][0][0]<= idade and idade<=escaloes[1][0][1]:
                    escaloes[1][1] += 1
            if idade>=25:
                for escalao in escaloes:
                    if idade >= escalao[0][0] and idade <= escalao[0][1]:
                        escalao[1] += 1
    return escaloes


def distrib_Glucose(lista):
    glicose_Max=0
    for i in lista:
        if int(i[-2]) > glicose_Max:
            glicose_Max = int(i[-2])
    while glicose_Max % 10 != 0:
        glicose_Max += 1   

    glicose_Min=glicose_Max
    for i in lista:
        if int(i[-2])<glicose_Min:
            glicose_Min=int(i[-2])
    while glicose_Min%10 !=0:
        glicose_Min -= 1
    
    escaloes = [[(x, x+9), 0] for x in range(glicose_Min, glicose_Max, 10)]
    for i in lista:
        if i[-1] == "1":
            glicose= int(i[-2])
            for escalao in escaloes:
                    if glicose >= escalao[0][0] and glicose <= escalao[0][1]:
                        escalao[1] += 1
    return escaloes

def tabela(dados,critério):
    if critério==distrib_Sexo:
        print(" Sexo  | Nº de pessoas doentes")
    if critério==distrib_Idades:
        print("Distribuição etária     | Nº de pessoas doentes")
    if critério==distrib_Glucose:
        print("Diabetes | Nº de doentes")
    for i in critério(dados):
        print(i[0], "|", i[1])


def menu():
    print("""Bem-vindo à análise sobre a diabetes! 
Menu:
    (1) Consultar distribuição por sexo
    (2) Consultar distribuição por faixa etária 
    (3) Consultar distribuição por nível de glicose no sangue
    (0) Sair """)
    return

menu()
function=int(input("O que pretende fazer?"))
print("")

dados=lerFicheiro("diabetes_prediction_dataset.csv")
while function !=0:
    if function == 1:
        tabela(dados[1:],distrib_Sexo)
        print("")

    if function == 2:
        tabela(dados[1:],distrib_Idades)
        print("")

    if function == 3:
       tabela(dados[1:],distrib_Glucose)
       print("")

    function = int(input("O que pretende consultar de seguida?"))
print("")
print("Saiu do programa da análise da diabetes! Volte sempre.")
print("")
```
