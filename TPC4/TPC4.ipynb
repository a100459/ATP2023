```python
import random

# Criar lista
def criaLista(N):
    i=0
    lista=[]
    while i <= N:
        elem = random.randint (1,101) 
        lista.append(elem)
        i = i+1
    print (lista) 
    return lista

# Ler lista
def leLista(N):
    lista1=[]
    i = 0
    while i<N:
        i = i+1
        lista1.append(int(input("Introduza um número inteiro")))
    print(lista1)
    return lista1

# Soma
def somaLista(lista):
    soma=0
    for n in lista:
        soma = soma + n   
    return soma

# Média
def mediaLista(lista):
    media=0
    for n in lista:
        media = somaLista(lista) / len(lista)
        n=n+1      
    return media

# Maior
def maiorLista(lista):
    maior=lista[0]
    for n in lista:
        if n>maior:
            maior = n
    return maior

# Menor
def menorLista(lista):
    menor=lista[0]
    for n in lista:
        if n<menor:
            menor = n
    return menor

# Ordem crescente
def estaOrdenadaC(lista):
    valor=True
    l=1 
    while l<len(lista):
        if lista[l] < lista[l-1]:
            valor=False
        l=l+1
    return valor

# Ordem decrescente
def estaOrdenadaD(lista):
    valor=True
    l=1
    while l<len(lista):
        if lista[l] > lista[l-1]:
            valor=False
        l=l+1
    return valor

# Procurar um elemento e devolver a sua posição
def indiceDe(lista, elem):
    res = -1
    i = 0
    procura = []
    while i < len(lista):
        if lista[i] == elem:
            res = i
            procura.append(i)
            print(procura)
        i = i+1
    return res

# Quadrados dos elementos da lista
def quadrados(lista):
    return[x*x for x in lista]


# Menu
def menu ():
    print(''' Menu:
(1) Criar Lista aleatória
(2) Ler Lista
(3) Soma
(4) Média
(5) Maior
(6) Menor
(7) estaOrdenada por ordem crescente
(8) estaOrdenada por ordem decrescente
(9) Procura um elemento
(10) Quadrados dos elementos
(0) Sair''')
    return


menu()
ordem = int(input("O que pretende selecionar no menu?"))
final = []
while ordem != 0:
    if ordem == 1:
        final.clear()
        final = criaLista(int(input("Selecione o tamanho da lista que pretende.")))

    if ordem == 2:
        final.clear()
        final = leLista(int(input("Introduza o tamanho da lista que pretende.")))

    if ordem == 3:
        print("A soma final dos elementos é de", somaLista(final),".")

    if ordem == 4:
        print("A média dos elementos é de", mediaLista(final),".")

    if ordem == 5:
        print("O maior elemento da lista é o", maiorLista(final),".")

    if ordem == 6:
        print("O menor elemento da lista é o", menorLista(final),".")

    if ordem == 7:
        if estaOrdenadaC(final)== True:
            print("Sim!")
        else:
            print("Não!")

    if ordem == 8:
        if estaOrdenadaD(final)== True:
            print("Sim!")
        else:
            print("Não!")

    if ordem == 9:
        procura = indiceDe(final,int(input("Qual é o elemento que procura?")))
        if procura ==-1:
            print("O elemento que procura não se encontra na lista.")
        else:
            print("o elemento encontra-se na",procura,"posição.")

    if ordem == 10:
        print("A lista dos quadrados dos elementos é ", quadrados(final),".")

    ordem=int(input("o que pretende selecionar?"))

print("a lista é:",final)

```
